<p style="text-align: justify; font-family: cursive; font-size: 14px; color: black;">
The manufacturing dataset consists of multiple process-related parameters that influence the
overall quality of the manufactured product. Each column represents a key physical or derived
metric used to model and predict product quality.
</p>

<ul style="font-family: cursive; font-size: 14px; color: black;">

  <li>
    <b>Temperature (°C):</b>
    Represents the operating temperature during the manufacturing process, measured in degrees
    Celsius. Temperature plays a critical role in determining material behavior, structural
    properties, and final product quality.
  </li>

  <li>
    <b>Pressure (kPa):</b>
    Indicates the pressure applied during the manufacturing process, measured in kilopascals (kPa).
    Pressure directly impacts material deformation, bonding, and transformation outcomes.
  </li>

  <li>
    <b>Temperature × Pressure:</b>
    An interaction feature capturing the combined effect of temperature and pressure. This term
    helps the model learn non-linear relationships arising from simultaneous changes in both
    parameters.
  </li>

  <li>
    <b>Material Fusion Metric:</b>
    A derived feature calculated as the square of temperature plus the cube of pressure.
    This metric represents material fusion behavior occurring during the manufacturing process.
  </li>

  <li>
    <b>Material Transformation Metric:</b>
    A derived feature computed as the cube of temperature minus the square of pressure.
    It provides insight into complex material transformation dynamics under varying process
    conditions.
  </li>

  <li>
    <b>Quality Rating:</b>
    The target variable representing the overall quality of the produced items.
    This rating serves as a key indicator of manufacturing performance and product reliability.
  </li>

</ul>

```python
def ZF_RF_Regression(X_train, X_test, y_train, y_test, n_estimators=200, max_depth=None, random_state=42):
    """
    Random Forest Regression Model
    Includes full evaluation and visualization similar to ANN function.
    """

    # ----------------------------
    # 1️⃣ Build and train Random Forest model
    model_rf = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=random_state,
        n_jobs=-1
    )
    model_rf.fit(X_train, y_train)

    # ----------------------------
    # 2️⃣ Predictions
    y_train_pred = model_rf.predict(X_train)
    y_test_pred = model_rf.predict(X_test)

    # ----------------------------
    # 3️⃣ Metrics
    metrics_train = {"RMSE": np.sqrt(mean_squared_error(y_train, y_train_pred)),
                     "MAE": mean_absolute_error(y_train, y_train_pred),
                     "R2": r2_score(y_train, y_train_pred)}
    metrics_test = {"RMSE": np.sqrt(mean_squared_error(y_test, y_test_pred)),
                    "MAE": mean_absolute_error(y_test, y_test_pred),
                    "R2": r2_score(y_test, y_test_pred)}

    print("=== RF Train Metrics ===", metrics_train)
    print("=== RF Test Metrics ===", metrics_test)

    sq_diff = np.mean((y_train_pred.mean() - y_test_pred.mean())**2)
    print(f"Train/Test Mean Squared Difference: {sq_diff:.4f}")

    # ----------------------------
    # Residuals
    residuals = y_test - y_test_pred
    standardized_residuals = residuals / np.std(residuals)

    # Residual plot
    plt.figure(figsize=(6, 5))
    plt.scatter(y_test_pred, residuals, alpha=0.6)
    plt.axhline(0, color="red", linestyle="--")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.title("Residual Plot (RF)")
    plt.show()

    # Standardized Residuals
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred, y=standardized_residuals, color='purple', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--')
    plt.axhline(2, color='red', linestyle='--')
    plt.axhline(-2, color='red', linestyle='--')
    plt.title("Standardized Residuals (RF)")
    plt.show()

    # ----------------------------
    # Predicted vs Actual (linear)
    plt.figure(figsize=(8, 6))
    max_val = max(y_test.max(), y_test_pred.max())
    plt.scatter(y_test, y_test_pred, color='red', alpha=0.6, label='Predicted')
    plt.plot([0, max_val], [0, max_val], color='green', linestyle='--', label='Perfect Prediction')
    plt.xlabel("Actual Hydrogen Flow Rate")
    plt.ylabel("Predicted Hydrogen Flow Rate")
    plt.title("RF: Predicted vs Actual")
    plt.legend()
    plt.show()

    # ----------------------------
    # Q-Q plot
    plt.figure(figsize=(8, 6))
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title('Q-Q Plot of Residuals (RF)')
    plt.show()

    # ----------------------------
    # Residual outliers
    threshold = 2 * np.std(residuals)
    outliers = residuals[np.abs(residuals) > threshold]
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred, y=residuals, color='blue', alpha=0.6, label='Residuals')
    plt.scatter(y_test_pred[np.abs(residuals) > threshold], outliers, color='red', label='Outliers')
    plt.axhline(0, color='black', linestyle='--')
    plt.title("Residuals Highlighting Outliers (RF)")
    plt.xlabel("Predicted Values")
    plt.ylabel("Residuals")
    plt.legend()
    plt.show()

    return model_rf, metrics_train, metrics_test, sq_diff
```

```python


def ZF_DT_Regression(X_train, X_test, y_train, y_test, max_depth=None, random_state=42):
    """
    Decision Tree Regression with full evaluation and visualization.
    """

    # ----------------------------
    # 1️⃣ Build Decision Tree model
    model_dt1 = DecisionTreeRegressor(max_depth=max_depth, random_state=random_state)

    # ----------------------------
    # 2️⃣ Train model
    model_dt1.fit(X_train, y_train.ravel())

    # ----------------------------
    # 3️⃣ Predictions
    y_train_pred1 = model_dt1.predict(X_train)
    y_test_pred1 = model_dt.predict(X_test)

    # ----------------------------
    # 4️⃣ Metrics
    metrics_train1 = {
        "RMSE": np.sqrt(mean_squared_error(y_train, y_train_pred1)),
        "MAE": mean_absolute_error(y_train, y_train_pred1),
        "R2": r2_score(y_train, y_train_pred1)
    }
    metrics_test1 = {
        "RMSE": np.sqrt(mean_squared_error(y_test, y_test_pred1)),
        "MAE": mean_absolute_error(y_test, y_test_pred1),
        "R2": r2_score(y_test, y_test_pred1)
    }

    print("=== Decision Tree Train Metrics ===", metrics_train1)
    print("=== Decision Tree Test Metrics ===", metrics_test1)

    sq_diff = np.mean((y_train_pred1.mean() - y_test_pred1.mean())**2)
    print(f"Train/Test Mean Squared Difference: {sq_diff:.4f}")

    # ----------------------------
    # 5️⃣ Residuals
    residuals = y_test - y_test_pred1
    standardized_residuals = residuals / np.std(residuals)

    plt.figure(figsize=(6, 5))
    plt.scatter(y_test_pred1, residuals, alpha=0.6)
    plt.axhline(0, color="red", linestyle="--")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.title("Residual Plot (Decision Tree)")
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred1, y=standardized_residuals, color='purple', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--')
    plt.axhline(2, color='red', linestyle='--')
    plt.axhline(-2, color='red', linestyle='--')
    plt.title("Standardized Residuals (Decision Tree)")
    plt.show()

    # ----------------------------
    # 6️⃣ Predicted vs Actual (linear)
    plt.figure(figsize=(8, 6))
    max_val = max(y_test.max(), y_test_pred1.max())
    plt.scatter(y_test, y_test_pred1, color='red', alpha=0.6, label='Predicted')
    plt.plot([0, max_val], [0, max_val], color='green', linestyle='--', label='Perfect Prediction')
    plt.xlabel("Actual Hydrogen Flow Rate")
    plt.ylabel("Predicted Hydrogen Flow Rate")
    plt.title("Decision Tree: Predicted vs Actual")
    plt.legend()
    plt.show()

    # ----------------------------
    # 7️⃣ Q-Q plot
    plt.figure(figsize=(8, 6))
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title('Q-Q Plot of Residuals (Decision Tree)')
    plt.show()

    # ----------------------------
    # 8️⃣ Residual outliers
    threshold = 2 * np.std(residuals)
    outliers = residuals[np.abs(residuals) > threshold]
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred1, y=residuals, color='blue', alpha=0.6, label='Residuals')
    plt.scatter(y_test_pred1[np.abs(residuals) > threshold], outliers, color='red', label='Outliers')
    plt.axhline(0, color='black', linestyle='--')
    plt.title("Residuals Highlighting Outliers (Decision Tree)")
    plt.xlabel("Predicted Values")
    plt.ylabel("Residuals")
    plt.legend()
    plt.show()

    return model_dt, metrics_train, metrics_test, sq_diff
```

```python

def ZF_SVR_Regression(X_train, X_test, y_train, y_test,
                      kernel='rbf', C=1.0, epsilon=0.1, gamma='scale', random_state=42):

    """
    SVR Regression with full evaluation and visualization.
    """

    # ----------------------------
    # 1️⃣ Build SVR model pipeline (scaling + SVR)
    model_svr = Pipeline([
        ('scaler', StandardScaler()),
        ('svr', SVR(kernel=kernel, C=C, epsilon=epsilon, gamma=gamma))
    ])

    # ----------------------------
    # 2️⃣ Train model
    model_svr.fit(X_train, y_train.ravel())

    # ----------------------------
    # 3️⃣ Predictions
    y_train_pred = model_svr.predict(X_train)
    y_test_pred = model_svr.predict(X_test)

    # ----------------------------
    # 4️⃣ Metrics
    metrics_train = {
        "RMSE": np.sqrt(mean_squared_error(y_train, y_train_pred)),
        "MAE": mean_absolute_error(y_train, y_train_pred),
        "R2": r2_score(y_train, y_train_pred)
    }

    metrics_test = {
        "RMSE": np.sqrt(mean_squared_error(y_test, y_test_pred)),
        "MAE": mean_absolute_error(y_test, y_test_pred),
        "R2": r2_score(y_test, y_test_pred)
    }

    print("=== SVR Train Metrics ===", metrics_train)
    print("=== SVR Test Metrics ===", metrics_test)

    sq_diff = np.mean((y_train_pred.mean() - y_test_pred.mean())**2)
    print(f"Train/Test Mean Squared Difference: {sq_diff:.4f}")

    # ----------------------------
    # 5️⃣ Residuals
    residuals = y_test - y_test_pred
    standardized_residuals = residuals / np.std(residuals)

    plt.figure(figsize=(6, 5))
    plt.scatter(y_test_pred, residuals, alpha=0.6)
    plt.axhline(0, color="red", linestyle="--")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.title("Residual Plot (SVR)")
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred, y=standardized_residuals, color='purple', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--')
    plt.axhline(2, color='red', linestyle='--')
    plt.axhline(-2, color='red', linestyle='--')
    plt.title("Standardized Residuals (SVR)")
    plt.show()

    # ----------------------------
    # 6️⃣ Predicted vs Actual
    plt.figure(figsize=(8, 6))
    max_val = max(y_test.max(), y_test_pred.max())
    plt.scatter(y_test, y_test_pred, color='red', alpha=0.6, label='Predicted')
    plt.plot([0, max_val], [0, max_val], color='green', linestyle='--', label='Perfect Prediction')
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title("SVR: Predicted vs Actual")
    plt.legend()
    plt.show()

    # ----------------------------
    # 7️⃣ Q-Q plot
    plt.figure(figsize=(8, 6))
    stats.probplot(residuals, dist="norm", plot=plt)
    plt.title('Q-Q Plot of Residuals (SVR)')
    plt.show()

    # ----------------------------
    # 8️⃣ Residual outliers
    threshold = 2 * np.std(residuals)
    outliers = residuals[np.abs(residuals) > threshold]
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=y_test_pred, y=residuals, color='blue', alpha=0.6, label='Residuals')
    plt.scatter(y_test_pred[np.abs(residuals) > threshold], outliers, color='red', label='Outliers')
    plt.axhline(0, color='black', linestyle='--')
    plt.title("Residuals Highlighting Outliers (SVR)")
    plt.xlabel("Predicted Values")
    plt.ylabel("Residuals")
    plt.legend()
    plt.show()

    return model_svr, metrics_train, metrics_test, sq_diff
```
