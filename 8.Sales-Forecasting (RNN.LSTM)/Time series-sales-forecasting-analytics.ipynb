  <!-- --------------------- ARIMA SECTION -------------------- -->
  <h2 style="color:blue;">1. ARIMA Model</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • <b>ARIMA (Auto-Regressive Integrated Moving Average)</b> models short-term patterns and trends.<br>
    • Uses parameters <b>(p, d, q)</b> for lag dependency, differencing, and moving average smoothing.<br>
    • Best when data has <b>no strong seasonality</b> but clear trends.<br>
    • Benchmark model for comparing more advanced architectures.
  </p>
  <hr>

  <!-- --------------------- SARIMA SECTION -------------------- -->
  <h2 style="color:blue;">2. SARIMA Model</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • <b>SARIMA</b> extends ARIMA by capturing <b>seasonality</b> explicitly.<br>
    • Uses <b>(p, d, q)</b> and seasonal <b>(P, D, Q, m)</b> parameters.<br>
    • Suitable for weekly sales with annual seasonal cycles.<br>
    • Produces stable multi-step forecasts with confidence intervals.
  </p>
  <hr>

  <!-- --------------------- LSTM SECTION -------------------- -->
  <h2 style="color:blue;">3. LSTM (Long Short-Term Memory)</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • An advanced <b>deep learning RNN</b> capable of capturing long-range dependencies.<br>
    • Learns patterns through memory cells and gates.<br>
    • Requires sliding-window supervised learning (lag features).<br>
    • Performs extremely well when data has <b>complex nonlinear relationships</b>.
  </p>
  <hr>

  <!-- --------------------- RNN SIMPLE SECTION -------------------- -->
  <h2 style="color:blue;">4. RNN</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • A simpler recurrent model compared to LSTM.<br>
    • Models sequential patterns but struggles with long-term memory.<br>
    • Good baseline to compare with LSTM/GRU.<br>
    • Often shows higher error due to vanishing gradient issues.
  </p>
  <hr>

  <!-- --------------------- EXPONENTIAL SMOOTHING -------------------- -->
  <h2 style="color:blue;">5. Exponential Smoothing (Holt–Winters)</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • A classical time-series model handling <b>level, trend, and seasonality</b>.<br>
    • Supports both additive and multiplicative seasonal patterns.<br>
    • Great for smooth, stable forecasts with low computational cost.<br>
    • Often used as a baseline due to high interpretability.
  </p>
  <hr>

  <!-- --------------------- HYBRID MODELS -------------------- -->
  <h2 style="color:blue;">6. Hybrid SARIMA–XGBoost / ARIMA–ML Models</h2>
  <p style="font-size:140%; color:#556b2f; font-family:Comic Sans MS; line-height:1.4;">
    • Combines <b>SARIMA (linear)</b> with <b>XGBoost (nonlinear)</b> residual correction.<br>
    • Workflow:<br>
    &nbsp;&nbsp;1. SARIMA predicts baseline<br>
    &nbsp;&nbsp;2. Compute residuals = Actual – Forecast<br>
    &nbsp;&nbsp;3. XGBoost learns nonlinear corrections<br>
    &nbsp;&nbsp;4. Final forecast = SARIMA + XGBoost<br><br>
    • Hybrid models usually reduce RMSE significantly compared to single models.
  </p>
  <hr>

  <p style="text-align: justify; font-family: cursive; font-size: 14px; color: green; font-weight: bold;">
     📌 Relationship Between Promotion and Sales (Pearson Correlation Test)
  </p>

  <p style="text-align: justify;">
      To answer the question <strong>"Does promotion improve the sales?"</strong>, the Pearson correlation test is applied since both variables (<em>promotion</em> and <em>sales</em>) are numerical. The Pearson correlation coefficient measures the linear relationship between two continuous variables, with values ranging between -1 and +1.
  </p>

  <p style="text-align: justify; font-size: 14px; color: green; font-weight: bold;">
       🔍 Statistical Hypotheses
  </p>

  <ul style="font-family: cursive; font-size: 14px;">
      <li><strong>H0 (p &gt; 0.05):</strong> Promotion does not affect store sales.</li>
      <li><strong>H1 (p &lt; 0.05):</strong> Promotion does affect store sales.</li>
  </ul>

  <p style="text-align: justify;">
      Based on the analysis, the Pearson correlation coefficient is found to be 
      <strong>0.4279</strong> and the corresponding <strong>p-value is 0.0</strong>. Since the p-value is lower than 0.05, the null hypothesis <strong>is rejected</strong>.
  </p>

  <p style="text-align: justify;">
      Therefore, it can be concluded that <strong>promotion has a statistically significant relationship with store sales</strong>. This means that promotional activities are associated with an increase in sales performance.
  </p>

  <p style="font-size: 14px; color: green; font-weight: bold;">
      📊 Augmented Dickey-Fuller (ADF) Test
  </p>

  <p>
      The <strong>Augmented Dickey-Fuller (ADF) test</strong> is a statistical test used to determine whether a time series is 
      <strong>stationary or non-stationary</strong>. Stationarity is extremely important in many time series models such as ARIMA.
  </p>

  <p>
     The ADF test evaluates the <strong>null hypothesis</strong> that the time series contains a <em>unit root</em>, meaning 
     that it is <strong>non-stationary</strong>. The alternative hypothesis suggests the series <strong>is stationary</strong>.
  </p>

  <p>
     When running the ADF test, two values are mainly interpreted:
  </p>

  <ul style="font-family: cursive; font-size: 14px;">
      <li><strong>ADF Statistic</strong> – negative value, the more negative → stronger evidence of stationarity</li>
      <li><strong>p-value</strong> – if below <strong>0.05</strong>, reject the null hypothesis</li>
  </ul>

  <p>
     A low p-value indicates strong evidence against the presence of a unit root and therefore supports that 
     the <strong>time series is stationary</strong>.
  </p>

  <p style="font-size: 14px; color: green; font-weight: bold;">
      ✔ Key Insights
  </p>

  <ul style="font-family: cursive; font-size: 14px;">
      <li>Time series <strong>must be stationary</strong> before applying ARIMA</li>
      <li><strong>AIC optimization</strong> helps select the best ARIMA parameters</li>
      <li>ARIMA is effective for data with <strong>trend</strong> and <strong>seasonality</strong></li>
  </ul>

</div>

```python
# =========================
# 1. Fit Holt-Winters model
# =========================

# -------------------------------
# 3) Simple Exponential Smoothing (Level only)
# -------------------------------
ses_model = SimpleExpSmoothing(train, initialization_method="estimated").fit()
ses_forecast = ses_model.forecast(len(test))

# -------------------------------
# 4) Holt (Level + Trend)
# -------------------------------
holt_model = Holt(train, initialization_method="estimated").fit()
holt_forecast = holt_model.forecast(len(test))

# -------------------------------
# 5) Holt-Winters (Level + Trend + Seasonality)
# Seasonal weekly pattern assumed 52 weeks/year
# -------------------------------
hw_model = ExponentialSmoothing(
    train,
    trend='add',
    seasonal='add',
    seasonal_periods=52,
    initialization_method="estimated"
).fit()

hw_forecast = hw_model.forecast(len(test))

# -------------------------------
# 6) Evaluation
# -------------------------------
def evaluate_model(true, pred, name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    print(f"{name} -> MAE: {mae:.2f}, RMSE: {rmse:.2f}")

evaluate_model(test, ses_forecast, "SES")
evaluate_model(test, holt_forecast, "Holt")
evaluate_model(test, hw_forecast, "Holt-Winters")

# -------------------------------
# 7) Plot
# -------------------------------
plt.figure(figsize=(14,5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(ses_forecast, label='SES Forecast')
plt.plot(holt_forecast, label='Holt Forecast')
plt.plot(hw_forecast, label='Holt-Winters Forecast')
plt.legend()
plt.title('Weekly Sales Forecasting (SES / Holt / Holt-Winters)')
plt.show()
```
